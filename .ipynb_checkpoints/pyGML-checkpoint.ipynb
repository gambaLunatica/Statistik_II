## OLS Regresión Lineal
Ya que tenemos los datos preparados y visualizados, ahora aplicaremos una regresión lineal para modelar la relación entre el log del peso corporal y el log del peso del cerebro.

1. **Importar Libreerías**
   Vamos a usar statsmodels, una librería muy útil para hacer regresiones y análisis estadísticos.

```python
import statsmodels.api as sm
```

2. **Definir las variables dependiente e independiente**

```python
X = animalsdata['log_BodyWt']   # Variable independiente (input)
y = animalsdata['log_BrainWt']  # Variable dependiente (output)
```

3. Agregar la constante (intercepto β₀)

```python
X = sm.add_constant(X)  # añade una columna de unos para β₀
```

4. **Ajustar el modelo OLS (Mínimos Cuadrados Ordinarios)**

```python
model = sm.OLS(y, X)        # define el modelo
results = model.fit()       # ajusta el modelo con los datos
print(results.summary())    # muestra un resumen con todos los resultados
```

5. **Visualizar la recta ajustada**

```python
import matplotlib.pyplot as plt

# Graficar los datos
plt.plot(animalsdata['log_BodyWt'], animalsdata['log_BrainWt'], 'o', label='data')

# Graficar la recta de regresión
plt.plot(animalsdata['log_BodyWt'], results.predict(X), 'r-', label='OLS fit')

plt.title('Linear Regression (OLS): log brain weight vs log body weight')
plt.xlabel('log body weight')
plt.ylabel('log brain weight')
plt.legend()
plt.show()
```

```python
'''usage of statsmodels with patsy-formula description'''
import statsmodels.formula.api as smf  

'''fit linear Gaussian (default) model with Patsy formula'''
wfit = smf.glm(formula='log_BrainWt ~ log_BodyWt', data=animalsdata).fit()
```

```python
'''extract parameters and values'''
b, m = wfit.params                                   # beta parameters: intercept and gain
print('regression            y  =     {:.3f}*x   +    {:.3f}   +  epsilon'.format(m, b))   
cib = (wfit.conf_int().iloc[0][0], wfit.conf_int().iloc[1][0]) # 1st ci[lower, upper] of intercept
cim = (wfit.conf_int().iloc[0][1], wfit.conf_int().iloc[1][1]) # 2nd ci[lower, upper] of gain
print('regression parameters CI = {}, {}'            #
      .format(np.round(cim, decimals=3), np.round(cib, decimals=3))) 
print('deviations epsilon ~ N({:.5f}, {:.5f})'       # distribution of residuals
      .format(wfit.resid_response.mean(), wfit.resid_response.var(ddof=1+wfit.df_model))) 
```

```python
'''fit linear model to data: confidence intervals of prediction and point estimate'''
# get data
x = np.asarray(animalsdata['log_BodyWt'])            # extract independent variable
y = np.asarray(animalsdata['log_BrainWt'])           # extract dependent variable
y_hat = np.asarray(wfit.fittedvalues)                # estimated result

fig = plt.figure(figsize=(6,4))
# show data
plt.plot(x, y, 'bo', markersize=8, markerfacecolor='None')
# show fit
plt.plot(x, y_hat,'-', color='0.1', linewidth=2, alpha=0.5, label='Fit')  

# grid for straight plotting between tails of fitted line
x2 = np.linspace(np.min(x), np.max(x), 100)
y2 = np.linspace(np.min(y_hat), np.max(y_hat), 100)    # IF SLEEP: reverse due to decrease

# statistics
t_975 = stats.t.ppf(0.975, wfit.df_resid)   # used for CI and PI bands acc. to residual's degree-of-freedom
s_err = np.sqrt(np.sum(wfit.resid_response**2)/(wfit.df_resid))   # estimated standard deviation of the error

# confidence interval
x_bar = x.mean()
CI = t_975 * s_err * np.sqrt(     1/len(x) + (x2-x_bar)**2/np.sum((x-x_bar)**2))
plt.fill_between(x2, y2+CI, y2-CI, color='b', alpha=.1, edgecolor='none', label='95% Confidence Limits')

# prediction interval
PI = t_975 * s_err * np.sqrt( 1 + 1/len(x) + (x2-x_bar)**2/np.sum((x-x_bar)**2))   
plt.plot(x2, y2-PI, '--', color='0.5', label='95% Prediction Limits')
plt.plot(x2, y2+PI, '--', color='0.5')

# labels
plt.title('animals sleep data - linear model fit with CI')
plt.xlabel('log body weight')
plt.ylabel('log brain weight')
plt.xticks([-2, -1, 0, 1, 2, 3, 4], ['10g', '100g', '1kg', '10kg', '100kg', '1t', '10t']) # unit kg
plt.yticks([-1, 0, 1, 2, 3], ['0.1g', '1g', '10g', '100g', '1kg'])    # unit g
plt.legend(loc='upper left');
```

```python
### Qué está pasando aquí?
Estás ajustando el modelo:

$$
\log(\text{BrainWt}) = \beta_0 + \beta_1 \cdot \log(\text{BodyWt}) + \varepsilon
$$

Y estás viendo si existe una relación lineal entre ambas variables en escala logarítmica.
```

## Diseño de Matrices

1. ### **Matriz con *Numpy***

Cuando usas numpy, tú armas las matrices a mano, lo que te da control total, pero implica más trabajo y cuidado.

**Pros:**
- Muy rápido y eficiente.
- Control total de cómo se construyen las matrices.
- Útil si entiendes bien la teoría de los modelos.

**Contras:**
- No maneja automáticamente variables categóricas o fórmulas.
- Es más fácil cometer errores al preparar los datos.
- No tiene etiquetas de columnas, solo números.

```python
'''create a design matrix with numpy matrix functions'''
x1 = animalsdata['log_BodyWt']             # select data "log body weight"
print(type(x1), x1.shape)                  # data vector
#Matriz Numpy
X1 = np.vstack((np.ones(len(x1)), x1)).T   # tuple of N-vectors -> 2xN ; transpose -> Nx2-array
print(type(X1), X1.shape)                  # design matrix with intercept
print(X1[:5,:])                            # show its head
```

2. ### Matriz con *Patsy*
*patsy* te permite usar fórmulas estilo R, y él genera automáticamente la matriz de diseño (X) y la variable objetivo (y).

**Pros:**
- Muy cómodo para modelos estadísticos.
- Interpreta fórmulas estilo `y ~ x1 + x2 + C(categ)`.
- Convierte automáticamente variables categóricas a dummies.
- Devuelve DataFrames con nombres de columnas.

**Contras:**
- Más lento que NumPy.
- Menos control fino.
-Necesita que el data esté bien limpio.

```python
'''patsy automagically creates a design matrix for linear model sleep vs. body weight'''
y, X = patsy.dmatrices('log_BrainWt ~ log_BodyWt', animalsdata)
print(X[:6], '... design matrix X: intercept + independent data')
print(y[:6], '... response vector y: dependent data')
```

### 3. Calcular el estimador de puntos con la matriz
Del vector de parámetros de la matriz de diseño y el vector de datos
$$\mathbf{\widehat{\boldsymbol\beta}} = (\mathbf{X}^T\;\mathbf{X})^{-1}\mathbf{X}^T \;\mathbf{y}$$

```python
'''solve pseudo inverse for parameter estimation'''
beta_hat = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
beta_hat = beta_hat[:, 0]      # make the 2x1  column matrix just an (unspecified) vector
print('mathematical solution for estimate of beta: [{:.4f}  {:.4f}]'
      .format(beta_hat[0], beta_hat[1]))
```

```python
yguess = X.dot(beta_hat)     # matrix-vector-product y = X beta
for i in range(7):           # cycle through lines of y-guess vs. y-measured
    print('X*beta={:6.2f}, measured={:6.2f}'.format(yguess[i], y[i, 0]))
```

```python
'''show linear dependency for    log(body weight) ~ log(brain weight)'''
fig = plt.figure(figsize=(6,4))
# show data
plt.plot(animalsdata['log_BodyWt'], animalsdata['log_BrainWt'],'o', label='data',
         markeredgewidth=1,markeredgecolor='b',markerfacecolor='None')

def f(x, beta):
    '''regression line y = f(x) = beta0+beta1*x  with parameters beta[] at value x'''
    return beta[0] + beta[1]*x

xs = np.asarray([animalsdata['log_BodyWt'].min(), animalsdata['log_BodyWt'].max()])
ys = f(xs, beta_hat)
plt.plot(xs, ys, 'g-', label='regression')   # straight line through extreme points

# Labels
plt.title('animals sleep data')
plt.xticks([-2, 0, 2, 4], ['10g', '1kg', '100kg', '10t'])
plt.yticks([0, 1, 2, 3], ['1g', '10g', '100g', '1kg'])
plt.xlabel('log body weight')
plt.ylabel('log brain weight')
plt.legend(loc='upper left');
```
